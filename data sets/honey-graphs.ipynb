```python
#Load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
%matplotlib inline
import matplotlib.ticker as ticker
from matplotlib import ticker
```

```python
#Import Cleaned Bee CSV
honey_bee_data_df = pd.read_csv("cleanhoney.csv")
```

```python
#Show the data
honey_bee_data_df
```

```python
#Set index to year
honey_bee_data_df.set_index('year', inplace=True)
```

```python
honey_bee_data_df['Total Production']
```

```python
#Graph total honey production

plt.figure(figsize=(15,7))
plt.plot(honey_bee_data_df['Total Production'], color = 'g',label = "Total Honey Production")

plt.xticks(honey_bee_data_df.index, rotation='vertical')

plt.xlabel('Year')
plt.ylabel('Total Production in Pounds')
plt.title('Total US Honey Production 1998-2019', fontsize = 15)
plt.legend()
plt.savefig("totalhoneyproduction.png", bbox_inches = "tight", transparent=True)
plt.show()
```

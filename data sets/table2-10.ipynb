```python
#Import things
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import scipy.stats as st
uncleandata = pd.DataFrame(pd.read_csv("table 2-10.csv"))
```

```python
#create a clean data frame to work with
df = (uncleandata.rename(columns=uncleandata.iloc[1]).loc[2:]).dropna(axis=1,how='all').replace(["+","+%"],0)
df.dropna(inplace=True)
newdf=df.rename(columns = {'Sector/Source':'year'})
totals=newdf.loc[newdf['year'].isin(['Transportation','Electric Power Industry','Industry','Agriculture','Commercial','Residential'])].set_index('year').T
totals.drop(totals.tail(1).index,inplace=True)
totals.to_csv("totalpollutionbysector.csv")
```
